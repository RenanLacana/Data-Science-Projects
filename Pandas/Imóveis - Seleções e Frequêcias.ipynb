## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv("imoveis_residenciais.csv", sep = ";")
```

```python
dados.head()
```

## Seleção de somente os imóveis classificados com o tipo "apartamento"

```python
selecao = dados.Tipo == "Apartamento"
n1 = dados[selecao]
print(n1.shape[0])
n1
```

## Seleção de imóveis classificados com o tipo "Casa", "Cada de Condomínio" ou "Casa de Vila"

```python
selecao = (dados.Tipo == "Casa") | (dados.Tipo == "Casa de Condomínio") | (dados.Tipo == "Casa de Vila")
n2 = dados[selecao]
print(n2.shape[0])
n2.Tipo.drop_duplicates()
```

```python
n2
```

## Seleção de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites

```python
selecao = (dados.Area >= 60) & (dados.Area <= 100)
n3 = dados[selecao]
print(n3.shape[0])
n3
```

## Seleção de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$2000,00

```python
selecao = (dados.Quartos >= 4) & (dados.Valor < 2000)
n4 = dados[selecao]
n4.shape[0]
```

```python
n4.head()
```

## Conclusão

```python
print("N° de os imóveis classificados com o tipo 'apartamento' -> {}".format(n1.shape[0]))
print("N° de imóveis classificados com o tipo 'Casa', 'Cada de Condomínio' ou 'Casa de Vila' -> {}".format(n2.shape[0]))
print("N° de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {}".format(n3.shape[0]))
print("N° de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$2000,00 -> {}".format(n4.shape[0]))
```
