## Tratando condominio

```python
dados[dados.Condominio.isnull()].shape[0]
```

```python
selecao = (dados.Tipo == "Apartamento") & (dados.Condominio.isnull())
```

```python
dados = dados[~selecao]  # Operador "~" inverte o valor booleano
```

```python
dados.info()
```

```python
dados.fillna(value={'Condominio': 0, 'IPTU': 0}, inplace = True) #Imóveis que não são apartamentos preechidos com condomínio = 0
```
