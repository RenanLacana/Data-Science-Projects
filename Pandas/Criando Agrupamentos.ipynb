```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=";")
```

```python
dados
```

```python
dados['Valor'].mean()
```

```python
dados.Bairro.drop_duplicates().shape[0]
```

```python
bairros = dados.Bairro.drop_duplicates()
```

```python
grupo_bairro = dados.groupby("Bairro")
```

```python
type(grupo_bairro)
```

```python
for bairro, data in grupo_bairro:
    print("{} -> {}".format(bairro, data.Valor.mean()))
```

```python
grupo_bairro['Valor'].mean()
```

```python
grupo_bairro[['Valor', 'Condominio']].mean()
```

```python
grupo_bairro.Valor.describe()
```

```python
grupo_bairro.Valor.aggregate(['min', 'max','sum']).rename(columns = {'min': 'Mínimo', 'max':'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro.Valor.mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```

```python
fig = grupo_bairro.Valor.max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Máximo do Aluguel por Bairro', {'fontsize':22})
```

```python
fig = grupo_bairro.Quartos.max().plot.bar(color = 'blue')
fig.set_ylabel('Número de Quartos')
fig.set_title('Número de Quartos Máximo por Bairro', {'fontsize':22})
```

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto')
produtos.describe().round(2)
```

```python
stat = precos.groupby('Produto').agg(['mean','std','min','max']).round(2).rename(columns = {'mean': 'Média', 'std': 'Desvio Padrão', 
    'min': 'Mínimo', 'max': 'Máximo'})
```
